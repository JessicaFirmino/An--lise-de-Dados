# Carregando os arquivos csv

1- Carregando e lendo arquivos

```python
import pandas as pd
import matplotlib.pyplot as plt

tabela = pd.read_csv("desafioCursos.csv", sep=",")


display(tabela)
```

2- Calculando receita total gerada pela venda dos cursos.

```python
tabela['Receita'] = tabela['Quantidade de Vendas'] * tabela['Preço Unitário']
total_receita = tabela['Receita'].sum()

print(f"O valor total da receita gerada pelas vendas dos cursos: R$ {
      total_receita:.2f}")
```

3- Identificando o curso com o maior número de vendas.

```python
curso_mais_vendido = tabela.loc[tabela['Quantidade de Vendas'].idxmax()]


print("O curso mais vendido foi:")

print(curso_mais_vendido)
```

3- Visualizar a distribuição das vendas ao longo do tempo através de gráficos.

```python

tabela['Data'] = pd.to_datetime(tabela['Data'])

tabela = tabela.sort_values(by='Data')

plt.figure(figsize=(10, 6))

plt.plot(tabela['Data'], tabela['Quantidade de Vendas'],
         marker='o', linestyle='-')

plt.title('Distribuição das Vendas ao Longo do Tempo')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')
plt.xticks(rotation=45)
plt.grid(True)

plt.tight_layout()
plt.show()
```
